#### 15. How to add a border (filled with 0's) around an existing array?

By the previous cell way, or by using np.pad, we can solve this question.

```python
arr2 = np.full((5,5), 1)
np.pad(arr2, pad_width=1, mode='constant', constant_values=0)
```

```python
help(np.pad)
```

It has 4 common modes - constant (pads the center array with a value in a width), edge (repeat the edge values), reflect (mirrors the inside values), wrap (considers the array as infinitely repeated in all the directions and set the next repeating number of the array)
