#### 17. Create a 8x8 matrix and fill it with a checkerboard pattern

Let black = 0 and white = 1

```python
temp_1 = np.array([0, 1, 0, 1, 0, 1, 0, 1], dtype = 'int')
temp_2 = np.array([1 ,0, 1, 0, 1, 0, 1, 0], dtype = 'int')

ori = np.zeros((8, 8), dtype='int')
for i in range(8):
    if i % 2 == 0:
        ori[i] = temp_1
    else:
        ori[i] = temp_2

print(ori)
```

My version would be slow as it switches from C and python again and again, I have seen a version in online which uses slicing to do the job, which would be fast.

```python
original = np.zeros((8, 8), dtype='int')
original[1::2, ::2] = 1
original[::2, 1::2] = 1
print(original)
```

This version would be fast, as it directly uses low level languages here (The whole numpy is written with Python, C and Fortron for improving performance).

#### 20. Normalize a 5x5 random matrix 

```python
random_array = np.random.randint(1, 10000, (5, 5))
random_array
```

In np.random.randint(), the shape is given as a tuple, plus, there are start and end values provided. In np.random.rand(), the shape is given as two integers, without start and stop value.

Following is the normalization formula.



```python
min_val = random_array.min()
max_val = random_array.max()

normalized_random_array = (random_array - min_val) / (max_val - min_val)
normalized_random_array
```

#### 22. Given a 1D array, negate all elements which are between 3 and 8, in place.

#### 23. How to round away from zero a float array ?

```python
arr = np.array([1.2, 3.4, -3.22, -0.32])
mask1 = arr > 0
mask2 = arr < 0
arr[mask1] = np.ceil(arr[mask1])
arr[mask2] = -1 * (np.ceil(-1 * arr[mask2]))
arr
```

#### 26. How to get the dates of yesterday, today and tomorrow?

```python
today = np.datetime64('today', 'D')
today
```

```python
yesterday = today - np.timedelta64(1, 'D')
yesterday
```

```python
tomorrow = today + np.timedelta64(1, 'D')
tomorrow
```

datetime64 helps to get today's date. 'D' helps in setting the format as day. timedelta64 helps to find the difference between any two points in the time..
