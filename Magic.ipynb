Features are columns in the dataset that help the model to understand the pattern to predict the label.

Here, as the class is known, this learning is called Supervised learning..
In Unsupervised learning, there won't be any classes. The model's main part is grouping or clustering.
In Reinforcement learning, there is an agent (Not an agent in current Generative AI sense) which learns in interactive environment based on rewards and penalties.

Qualitative features describe a non numerical characteristic, category or label. They can't be measured with numbers but instead, a quality or category such as an item's color, etc. 

There are two types:
1. Nominal features - Categories with no logical order or hierarchy.
2. Ordinal features - Categories that have a meaningful order or ranking.

For non ordinal features, one hot encoding is a best choice. For eg, if there are three colours, say Red, Green and Blue, one hot encoding creates three extra columns such as is_red, is_green and is_blue. If the colour is red, the is_red column will become 1, and so on. Without this, if we encode Red as 1, Blue as 2 and Green as 3, then, ML models may create an unwanted linear hierarchical relationship between the columns, with which, the prediction might go wrong.

This may affect at the places where, the Linear models, Distance based models or Gradient based optimization might be used as, there is a possibility for those models to learn a different, unwanted, wrong relationship between the color encoding and the other features.

For ordinal categories such as bad, satisfactory, neutral, good and great, the encoding can be given as 1, 2, 3, 4 and 5. The Distance based models can analyse and understand how related Good and Bad are, etc.

Quantitative features describe a numeric characteristic, category or label. They represent numeric quantities. There are two types in that.

1. Discrete Quantitative Features - Countable quantities, Values are usually integers because you count individual items. Eg: Number of users, Number of packets transmitted.
2. Continuous Quantitative Features - Measurable quantities, Can take any value within a range, including decimals. Eg: Height, Weight, Temperature etc.

No encoding is needed here, as they are already Math.

In supervised learning, the Model can predict discrete classes. (Multiclass Classification, Binary classification)

For Binary Classification, Eg: Positive / negative, Cat / dog, Spam / not spam

For Multiclass Classification, Eg: Cat/dog/lizard/dolphin, Orange/apple/pear, Plant species

The Model can also predict Continuous values - Regression
Eg: Price of the house, Price of the asset, etc.

Feature vector - Features in one sample
Feature matrix - Features of all the samples

The whole dataset is split into Training dataset, Validation dataset and testing dataset (60%, 20%, 20% respectively or 80%, 10%, 10% anything)...

While using Training dataset, we'll make the model predict the output and check how close the prediction is aligned with the output. The difference between the prediction and the actual value is called Loss.
With the Loss, we'll make adjustments. So, in training, the feedback loop happens.

Then, we'll check whether the model can handle unseen data using the validation set. 

With the validation set, we'll evaluate which model is good on the comparison, with which, at the end, the model is tested with the Test set, where, the performance is analysed.

Loss - Difference between the Prediction and the actual label.

Some Loss functions:
1. L1 Loss = sum(|y_real - y_predicted|)
Here, sum is done to add the Loss values of each and every entry of the dataset and with that, the overall Loss is found.
2. L2 loss = sum((y_real - y_predicted) ^ 2)
3. Binary Cross Entropy Loss (A big formula)
- Here, when the performance increases, the loss decreases and vice versa.
