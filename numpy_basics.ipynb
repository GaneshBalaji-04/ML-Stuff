```python
a = np.array([1, 2, 3])
b = a.copy()
b[0] = 100
print(a)
print(b)
```

#### Linear Algebra

```python
a = np.ones((2, 3))
print(a)

b = np.full((3, 2), 2)
print(b)

c = np.matmul(a, b)

print(c)
```

This matmul is used to perform the matrix multiplication.

```python
c[0, 1] = 4
np.linalg.det(c)
```

This method is used to find the determinant of a square matrix.
