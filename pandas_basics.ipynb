df.sample helps to get some sample data from the dataset, by default, the number of rows returned is 1, we can pass number of rows to return.
```python
coffee.loc[3]
```

```python
coffee.loc[[1, 4, 5]]
```

```python
coffee.loc[4:8]
```

df.loc[] is a property, which, can be used to access, select and modify data in a dataframe using label-based indexing. Syntax: df.loc[row_indexer, column_indexer]. Above formats can also be used..

```python
coffee.loc[5:8, ["Day", "Units Sold"]]
```

```python
coffee.loc[:, ["Day", "Units Sold"]]
```

In loc, upper index is inclusive, which is not the case in iloc. In iloc, instead of the column names, we treat them as 0 based index, then, provide the index values as columns.

```python
coffee.iloc[5:8, [0, 2]]
```
